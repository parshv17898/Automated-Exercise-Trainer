### Pose.py

```
%%writefile pose.py
import numpy as np

class PoseSequence:
    def __init__(self, sequence):
        self.poses = []
        for parts in sequence:
            self.poses.append(Pose(parts))
        
        # normalize poses based on the average torso pixel length
        torso_lengths = np.array([Part.dist(pose.neck, pose.mhip) for pose in self.poses if pose.neck.exists and pose.lhip.exists])  

        mean_torso = np.mean(torso_lengths)

        for pose in self.poses:
            for attr, part in pose:
                setattr(pose, attr, part / mean_torso) 

class Pose:
    PART_NAMES = ['nose', 'neck',  'rshoulder', 'relbow', 'rwrist', 'lshoulder', 'lelbow', 'lwrist','mhip', 'rhip', 'rknee', 'rankle', 'lhip', 'lknee', 'lankle', 'reye', 'leye', 'rear', 'lear', 'lbigtoe', 'lsmalltoe', 'lheel', 'rbigtoe', 'rsmalltoe', 'rheel']
    
    def __init__(self, parts):
        """Construct a pose for one frame, given an array of parts
        Arguments:
            parts - 25 * 3 ndarray of x, y, confidence values
        """ 
        for name, vals in zip(self.PART_NAMES, parts):
            setattr(self, name, Part(vals))
    
    def __iter__(self):
        for attr, value in self.__dict__.items():
            yield attr, value
    
    def __str__(self):
        out = ""
        for name in self.PART_NAMES:
            _ = "{}: {},{}".format(name, getattr(self, name).x, getattr(self, name).x)
            out = out + _ +"\n"
        return out
    
    def print(self, parts):
        out = ""
        for name in parts:
            if not name in self.PART_NAMES:
                raise NameError(name)
            _ = "{}: {},{}".format(name, getattr(self, name).x, getattr(self, name).x)
            out = out + _ +"\n"
        return out

class Part:
    def __init__(self, vals):
        self.x = vals[0]
        self.y = vals[1]
        self.c = vals[2]
        self.exists = self.c != 0.0

    def __floordiv__(self, scalar):
        __truediv__(self, scalar)

    def __truediv__(self, scalar):
        return Part([self.x / scalar, self.y / scalar, self.c])

    @staticmethod
    def dist(part1, part2):
        return np.sqrt(np.square(part1.x - part2.x) + np.square(part1.y - part2.y))
```

```
%%writefile utils.py
import os
import sys
import json
import numpy as np

def split_num(s):
    head = s.rstrip('0123456789')
    tail = s[len(head):]
    return head, tail

def files_in_order(folderpath):

    total_npy = []

    for fol in os.listdir(folderpath):
        
        npy_files = os.listdir(folderpath + '/' + fol)

        no_extensions = [os.path.splitext(npy_file)[0] for npy_file in npy_files]

        splitted = [split_num(s) for s in no_extensions]

        splitted = np.array(splitted)
        
        indices = np.lexsort((splitted[:, 1].astype(int), splitted[:, 0]))

        npy_files = np.array(npy_files)[indices]

        total_npy.append(npy_files)

    total_npy = np.array(total_npy)

    li = []

    for i in total_npy:
        for j in i:
            li.append(j)

    return li

# Generates binary labels (good=1, bad=0) given an array-like of filenames
def get_labels(array):

    y = np.zeros((len(array),2))
    count = 0
    for i in array:
          y[count,0] = hash_ex(i)

          if "good" in i:
              y[count,1] = 1
          else:
              y[count,1] = 0

          count = count+1
    
    return np.array(y)

# Compute Dynamic Time Warp Distance of two sequences
# http://alexminnaar.com/time-series-classification-and-clustering-with-python.html

def DTWDistance(s1, s2):
    DTW={}

    for i in range(len(s1)):
        DTW[(i, -1)] = float('inf')
    for i in range(len(s2)):
        DTW[(-1, i)] = float('inf')
    DTW[(-1, -1)] = 0

    for i in range(len(s1)):
        for j in range(len(s2)):
            dist= (s1[i]-s2[j])**2
            DTW[(i, j)] = dist + min(DTW[(i-1, j)],DTW[(i, j-1)], DTW[(i-1, j-1)])

    return np.sqrt(DTW[len(s1)-1, len(s2)-1])
```

```
%%writefile evaluate.py
import os
import numpy as np
from fn_calls import feedback_method
from scipy.signal import medfilt


good_bad = ['hyper_extension', 'Backward_arm_circles', 'Backward_lunges_left', 'Backward_lunges_right', 'Clap_pushups', 'Diamond_pushups','Forward_arm_circles', 'Forward_lunges_left', 'Forward_lunges_right', 'Front_raise', 'High_jumps', 'High_plank_front_new','Jumping_jacks', 'Lateral_raise','Lunges_left','Lunges_right','Modified_pushups','Plank_side', 'Pushup_new', 'Scapular_pushups', 'Split_jumps', 'Squat_hold', 'Squat_jacks', 'Squat_jumps', 'Squats', 'Sumo_squats', 'Wall_chair', '180_squat_jumps', 'airplane_lunges', 'bulgarian_split_squat_left', 'bulgarian_split_squat_right', 'Curtsy_lunges', 'Squat_kicks', 'squat_with_lunges', 'Tiptoe_squats', 'Warrior_lunges_right', 'Warrior_lunges_left', 'Shoulder_tap_pushups', 'Burpees', 'staggered_pushups', 'tricep_dips', 'downward_dog_to_pushup', 'incline_pushups', 'star_jump_burpees', 'Jack_knife_crunches', 'Jack_knife', 'reverse_crunches', 'Full_crunches', 'Hip_thrusts', 'Bridge_hold', 'lateral_plank_walk', 'body_saw', 'forearm_plank_reach_out', 'high_plank_jacks', 'Plank_hold', 'Side_plank_hold', 'Leg_raises', 'russian_twist', 'Crunches']

right_left = ['Elbow_to_knee_front_bend', 'Forward_punches', 'Front_kicks', 'High_plank_arm_reaches', 'High_plank_shoulder_tap', 'Jog_in_place', 'Mountain_climber','Lateral_leg_raises', 'Stance_switches', 'Yoga_tree_pose', 'Quick_feet_in_and_out', 'High_knees', 'Plank_walk', 'Seated_straddle_reaches', 'Side_squats', 'Standing_gate_openers', 'Superman_reaches', 'T_pushups', 'bicycle_crunches', 'heel_taps', 'plank_with_spinal_rotation']


def evaluate_pose(pose_seq, exercise):
    """Evaluate a pose sequence for a particular exercise.
    Args:
        pose_seq: PoseSequence object.
        exercise: String name of the exercise to evaluate.
    Returns:
        correct: Bool whether exercise was performed correctly.
        feedback: Feedback string.
    """
    ex_dict = np.load("/content/drive/My Drive/Data/Infivolve/Project_GenEx/Ex_Keypoints_mapping.npy", allow_pickle = 'TRUE').item()
    feed_dict = np.load("/content/drive/My Drive/Data/Infivolve/Project_GenEx/Ex_keypoints_UpDown_feedback_mapping.npy",allow_pickle = 'TRUE').item()

    keypoints = ex_dict[exercise]

    key_list = str(keypoints).split(",")

    poses = pose_seq.poses
    right_present = [1 for pose in poses if pose.rshoulder.exists and pose.relbow.exists and pose.rwrist.exists and pose.rhip.exists and pose.rknee.exists and pose.rankle.exists]
    left_present = [1 for pose in poses 
                        if pose.lshoulder.exists and pose.lelbow.exists and pose.lwrist.exists and pose.lhip.exists and pose.lknee.exists and pose.lankle.exists]
    right_count = sum(right_present)
    left_count = sum(left_present)

    if right_count==0 and left_count==0:
      # return a proper feedback and audio file
      return ('Proper Key points are not present in Data !','', '0', '')
        
    side = 'Right' if right_count > left_count else 'Left'
            
    if side == 'Right': #pose.neck
      joints = [(pose.rshoulder, pose.relbow, pose.rwrist, pose.rhip, pose.rknee, pose.rankle) for pose in poses]
    else:
      joints = [(pose.lshoulder, pose.lelbow, pose.lwrist, pose.lhip, pose.lknee, pose.lankle) for pose in poses]

    # filter out data points where a part does not exist
    joints = [joint for joint in joints if all(part.exists for part in joint)]
    joints_ = np.array(joints)

    arr1_min = []
    arr1_max = []

    shoulder_elbow_vecs = np.array([(joint[0].x - joint[1].x, joint[0].y - joint[1].y) for joint in joints])
    elbow_wrist_vecs = np.array([(joint[1].x - joint[2].x, joint[1].y - joint[2].y) for joint in joints])
    shoulder_hip_vecs = np.array([(joint[0].x - joint[3].x, joint[0].y - joint[3].y) for joint in joints])
    hip_knee_vecs = np.array([(joint[3].x - joint[4].x, joint[3].y - joint[4].y) for joint in joints])
    knee_ankle_vecs = np.array([(joint[4].x - joint[5].x, joint[4].y - joint[5].y) for joint in joints])

    shoulder_elbow_vecs = shoulder_elbow_vecs / np.expand_dims(np.linalg.norm(shoulder_elbow_vecs, axis=1), axis=1)
    elbow_wrist_vecs = elbow_wrist_vecs / np.expand_dims(np.linalg.norm(elbow_wrist_vecs, axis=1), axis=1)
    shoulder_hip_vecs = shoulder_hip_vecs / np.expand_dims(np.linalg.norm(shoulder_hip_vecs, axis=1), axis=1)
    hip_knee_vecs = hip_knee_vecs / np.expand_dims(np.linalg.norm(hip_knee_vecs, axis=1), axis=1)
    knee_ankle_vecs = knee_ankle_vecs / np.expand_dims(np.linalg.norm(knee_ankle_vecs, axis=1), axis=1)
            
    shoulder_elbow_wrist_angle = np.degrees(np.arccos(np.clip(np.sum(np.multiply(shoulder_elbow_vecs, elbow_wrist_vecs), axis=1), -1.0, 1.0)))
    shoulder_elbow_wrist_angle_filtered = medfilt(medfilt(shoulder_elbow_wrist_angle, 5), 5)
            
    shoulder_elbow_wrist_angle_min = np.min(shoulder_elbow_wrist_angle_filtered)#np.quantile(shoulder_elbow_wrist_angle_filtered,0.1)
    shoulder_elbow_wrist_angle_max = np.max(shoulder_elbow_wrist_angle_filtered)#np.quantile(shoulder_elbow_wrist_angle_filtered,0.9)

    arr1_min.append(np.floor(shoulder_elbow_wrist_angle_min))
    arr1_max.append(np.ceil(shoulder_elbow_wrist_angle_max))
            
    shoulder_elbow_hip_angle = np.degrees(np.arccos(np.clip(np.sum(np.multiply(shoulder_elbow_vecs, shoulder_hip_vecs), axis=1), -1.0, 1.0)))
    shoulder_elbow_hip_angle_filtered = medfilt(medfilt(shoulder_elbow_hip_angle, 5), 5)
            
    shoulder_elbow_hip_angle_min = np.min(shoulder_elbow_hip_angle_filtered)#np.quantile(shoulder_elbow_hip_angle_filtered,0.1)
    shoulder_elbow_hip_angle_max = np.max(shoulder_elbow_hip_angle_filtered)#np.quantile(shoulder_elbow_hip_angle_filtered,0.9)

    arr1_min.append(np.floor(shoulder_elbow_hip_angle_min))
    arr1_max.append(np.ceil(shoulder_elbow_hip_angle_max))
            
    shoulder_hip_knee_angle = np.degrees(np.arccos(np.clip(np.sum(np.multiply(shoulder_hip_vecs, hip_knee_vecs), axis=1), -1.0, 1.0)))
    shoulder_hip_knee_angle_filtered = medfilt(medfilt(shoulder_hip_knee_angle, 5), 5)
            
    shoulder_hip_knee_angle_min = np.min(shoulder_hip_knee_angle_filtered)#np.quantile(shoulder_hip_knee_angle_filtered,0.1)
    shoulder_hip_knee_angle_max = np.max(shoulder_hip_knee_angle_filtered)#np.quantile(shoulder_hip_knee_angle_filtered,0.9)

    arr1_min.append(np.floor(shoulder_hip_knee_angle_min))
    arr1_max.append(np.ceil(shoulder_hip_knee_angle_max))
            
    hip_knee_ankle_angle = np.degrees(np.arccos(np.clip(np.sum(np.multiply(hip_knee_vecs, knee_ankle_vecs), axis=1), -1.0, 1.0)))
    hip_knee_ankle_angle_filtered = medfilt(medfilt(hip_knee_ankle_angle, 5), 5)
            
    hip_knee_ankle_angle_min = np.min(hip_knee_ankle_angle_filtered)#np.quantile(hip_knee_ankle_angle_filtered,0.1)
    hip_knee_ankle_angle_max = np.max(hip_knee_ankle_angle_filtered)#np.quantile(hip_knee_ankle_angle_filtered,0.9)

    arr1_min.append(np.floor(hip_knee_ankle_angle_min))
    arr1_max.append(np.ceil(hip_knee_ankle_angle_max))
        
    arr2_min=[]
    arr2_max=[]

    if exercise in good_bad:
        good_file = np.load('Angle_npy/'+exercise+'/'+exercise+'_Good.npy')
    else:
        good_file = np.load('Angle_npy/'+exercise+'/'+side+'/'+exercise+'_Good.npy')

    for keys in range(4):
      arr2_max.append(np.ceil(np.quantile(good_file,0.75, axis = 0)[int(keys)])) #0.75
      arr2_min.append(np.floor(np.quantile(good_file,0.25, axis = 0)[int(keys)+4])) #0.25
      
    arr1_min = np.array(arr1_min)
    #print("\ntest_min_Important_angle : ")
    #print(arr1_min)
    arr1_max = np.array(arr1_max)
    #print("test_max_Important_angle : ")
    #print(arr1_max)
    arr2_min = np.array(arr2_min)
    #print("Ideal_min_Important_angle : ")
    #print(arr2_min)
    arr2_max = np.array(arr2_max)
    #print("Ideal_max_Important_angle : ")
    #print(arr2_max)
    
    return feedback_method(arr1_min, arr1_max, arr2_min, arr2_max, key_list, feed_dict, exercise)
```

### fn_calls.py

```
%%writefile fn_calls.py
import os
import numpy as np

def feedback_method(arr1_min, arr1_max, arr2_min, arr2_max, key_list, feed_dict, exercise):
    count = 0
    DETECTED = True
    correct = True
    feedback = ""
    strictness = 5

    for keys in key_list:
        if arr1_min[int(keys)]<max((arr2_min[int(keys)]-strictness),0):
            correct = False
            try:
                # print(feed_dict[exercise, int(keys), 'D1'])
                feedback = feed_dict[exercise, int(keys), 'D1'][0]
                audio_path = feed_dict[exercise, int(keys), 'D1'][1]
                return (correct, feedback, DETECTED, audio_path)
            except:
                feedback = ""

        if arr1_min[int(keys)]>(arr2_min[int(keys)]+strictness):
            correct = False
            try:
                # print(feed_dict[exercise, int(keys), 'D2'])
                feedback = feed_dict[exercise, int(keys), 'D2'][0]
                audio_path = feed_dict[exercise, int(keys), 'D2'][1]
                return (correct, feedback, DETECTED, audio_path)
            except:
                feedback = ""

        if arr1_max[int(keys)]>(arr2_max[int(keys)]+strictness):
            correct = False
            try:
                # print(feed_dict[exercise, int(keys), 'U1'])
                feedback = feed_dict[exercise, int(keys), 'U1'][0]
                audio_path = feed_dict[exercise, int(keys), 'U1'][1]
                return (correct, feedback, DETECTED, audio_path)
            except:
                feedback = ""

        if arr1_max[int(keys)]<max((arr2_max[int(keys)]-strictness),0):
            correct = False
            try:
                # print(feed_dict[exercise, int(keys), 'U2'])
                feedback = feed_dict[exercise, int(keys), 'U2'][0]
                audio_path = feed_dict[exercise, int(keys), 'U2'][1]
                return (correct, feedback, DETECTED, audio_path)
            except:
                feedback = ""

    if feedback == "":
        correct = True
        feedback = 'Thats good form, keep going!'
        audio_path = '0.mp3'
        return (correct, feedback, DETECTED, audio_path)
```
